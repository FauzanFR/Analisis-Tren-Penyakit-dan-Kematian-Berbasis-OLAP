# TRANSFORM

Setelah pengecekan, ditemukan:
- Nilai kosong di Mortality (22), Life Expectancy (1), Population (1), Income Level (20-29)
- Inkonsistensi bahasa antara nama negara Inggris vs Indonesia
- Jenis kelamin tidak baku (tidak_diketahui) dan level pendapatan Not classified

Semua telah dibersihkan: drop missing, mapping negara via googletrans + fuzzy, dan kategorisasi ulang.
Hasil: 0 missing value, 0 duplikat di semua tabel.


## Pembersihan data


```python

print("Mortality:")
n0 = len(df_mortality)
df_mortality = df_mortality.dropna(subset=['2017', '2018', '2019', '2020'], how='all').reset_index(drop=True)
print(f"  Drop missing: {n0} -> {len(df_mortality)} baris ({n0 - len(df_mortality)} dihapus)")

print("\nLife expectancy:")
n0 = len(df_life_expectancy)
df_life_expectancy = df_life_expectancy.dropna(subset=['2017', '2018', '2019', '2020'], how='all').reset_index(drop=True)
print(f"  Drop missing: {n0} -> {len(df_life_expectancy)} baris ({n0 - len(df_life_expectancy)} dihapus)")

print("\nPopulation:")
n0 = len(df_population)
df_population = df_population.dropna(subset=['2017', '2018', '2019', '2020'], how='all').reset_index(drop=True)
print(f"  Drop missing: {n0} -> {len(df_population)} baris ({n0 - len(df_population)} dihapus)")
```

```python
print("Income level - drop NaN:")
n0 = len(df_income_level)
df_income_level = df_income_level.dropna(subset=['kode_negara']).reset_index(drop=True)
print(f"  Drop: {n0} -> {len(df_income_level)} baris ({n0 - len(df_income_level)} dihapus)")

print("\nIsi NaN gdp_per_capita:")
n_null = df_income_level['gdp_per_capita'].isna().sum()
median_gdp = df_income_level.groupby('level_pendapatan')['gdp_per_capita'].transform('median')
df_income_level['gdp_per_capita'] = df_income_level['gdp_per_capita'].fillna(median_gdp)
# sisanya yang masih NaN (karena groupnya semua NaN) diisi median global
df_income_level['gdp_per_capita'] = df_income_level['gdp_per_capita'].fillna(df_income_level['gdp_per_capita'].median())
print(f"  {n_null} baris diisi dengan median per level_pendapatan")

# 5. Income Level — Kategorikan Not classified -> Unknown
print("\nLevel pendapatan:")
print(f"  Sebelum: {df_income_level['level_pendapatan'].unique()}")
df_income_level['level_pendapatan'] = df_income_level['level_pendapatan'].replace('Not classified', 'Unknown')
n_lvl_null = df_income_level['level_pendapatan'].isna().sum()
df_income_level = df_income_level.dropna(subset=['level_pendapatan']).reset_index(drop=True)
print(f"  Sesudah: {df_income_level['level_pendapatan'].unique()}")
print(f"  {n_lvl_null} baris dengan level_pendapatan NaN dihapus")
print(f"  Total akhir: {len(df_income_level)} baris")
```

```python
# 6. Mapping negara Cause of Death (Indonesia -> Inggris -> kode_negara)
from difflib import SequenceMatcher as SM
from googletrans import Translator

print("MAPPING NEGARA CAUSE OF DEATH :")

id_names = sorted(df_cause_of_death['negara'].unique())
en_names = dict(zip(df_mortality['negara'].str.lower(), df_mortality['negara']))
en_names.update(dict(zip(df_life_expectancy['negara'].str.lower(), df_life_expectancy['negara'])))
en_code = dict(zip(df_mortality['negara'], df_mortality['kode_negara']))
en_code.update(dict(zip(df_life_expectancy['negara'], df_life_expectancy['kode_negara'])))

print(f"  Negara unik di Cause of Death: {len(id_names)}")
print(f"  Referensi nama Inggris: {len(en_names)}")

# Override terjemahan tidak akurat
manual_override = {
    'Belanda': 'Netherlands',
    'Prancis': 'France',
    'Spanyol': 'Spain',
    'Rusia': 'Russian Federation',
    'Korea selatan': 'Korea, Rep.',
    'Kirgizstan': 'Kyrgyz Republic',
    'Mesir': 'Egypt, Arab Rep.',
    'Serbia dan montenegro': 'Serbia',
    'Hong kong (cina)': 'Hong Kong SAR, China',
    'Antilen belanda': 'Curacao',
    'Guyana prancis': 'France',
    'Guadeloupe': 'France',
    'Martinique': 'France',
    'Reunion': 'France',
    'Montserrat': 'United Kingdom',
    'Wallis dan futuna': 'France',
}

translator = Translator()
mapping = {}
unmapped = []

for id_name in id_names:
    if id_name in manual_override:
        target = manual_override[id_name]
        if target in en_code:
            mapping[id_name] = (target, en_code[target])
            continue
    
    try:
        translated = translator.translate(id_name, src='id', dest='en').text
    except:
        translated = id_name
    
    en_key = translated.strip().lower()
    
    if en_key in en_names:
        real_en = en_names[en_key]
        mapping[id_name] = (real_en, en_code[real_en])
    else:
        best_ratio = 0
        best_match = None
        for candidate_key, candidate_val in en_names.items():
            ratio = SM(None, en_key, candidate_key).ratio()
            if ratio > best_ratio:
                best_ratio = ratio
                best_match = candidate_val
        
        if best_ratio >= 0.6:
            mapping[id_name] = (best_match, en_code[best_match])
        else:
            unmapped.append((id_name, translated, best_match if best_match else '-', round(best_ratio, 2)))

print(f"  Berhasil di-mapping otomatis: {len(mapping)}/{len(id_names)} negara")

if unmapped:
    print(f"\n  Tidak ter-mapping ({len(unmapped)} negara):")
    for id_name, trans, best, score in unmapped:
        print(f"  {id_name} -> {best} (skor: {score})")

df_cause_of_death['negara_inggris'] = df_cause_of_death['negara'].map(lambda x: mapping[x][0] if x in mapping else x)
df_cause_of_death['kode_negara'] = df_cause_of_death['negara'].map(lambda x: mapping[x][1] if x in mapping else None)

n_unmapped_final = df_cause_of_death['kode_negara'].isna().sum()
print(f"\n  Final: {n_unmapped_final} baris tanpa kode_negara (akan di-drop)")
df_cause_of_death = df_cause_of_death.dropna(subset=['kode_negara']).reset_index(drop=True)
df_cause_of_death['kode_negara'] = df_cause_of_death['kode_negara'].astype(str)
print(f"  Setelah drop: {len(df_cause_of_death)} baris")
```

```python
print('Jenis kelamin:')
print(f"  Sebelum: {df_cause_of_death['jenis_kelamin'].unique()}")
df_cause_of_death['jenis_kelamin'] = df_cause_of_death['jenis_kelamin'].replace('tidak_diketahui', 'Unknown')
print(f"  Sesudah: {df_cause_of_death['jenis_kelamin'].unique()}")
```

## Star schema


```python
YEARS = ['2017', '2018', '2019', '2020']

def unpivot(df, value_name):
    melted = df.melt(
        id_vars=['kode_negara', 'negara'],
        value_vars=YEARS,
        var_name='tahun',
        value_name=value_name
    )
    melted['tahun'] = melted['tahun'].astype(int)
    return melted

print('Unpivot wide ke long:')

long_mortality = unpivot(df_mortality, 'mortality_rate')
print(f'  Mortality unpivot: {len(long_mortality)} baris')

long_life = unpivot(df_life_expectancy, 'life_expectancy')
print(f'  Life Expectancy unpivot: {len(long_life)} baris')

long_pop = unpivot(df_population, 'population')
print(f'  Population unpivot: {len(long_pop)} baris')
```

```python

# DimCountry — union semua negara
df_all_country = pd.concat([
    long_mortality[['kode_negara', 'negara']],
    long_life[['kode_negara', 'negara']],
    long_pop[['kode_negara', 'negara']],
    df_income_level[['kode_negara', 'negara']],
]).drop_duplicates('kode_negara').reset_index(drop=True)
df_dim_country = df_all_country.copy()
n = len(df_dim_country)
df_dim_country['country_id'] = [f'C{str(i).zfill(3)}' for i in range(1, n + 1)]
df_dim_country = df_dim_country[['country_id', 'kode_negara', 'negara']]
print(f'DimCountry: {len(df_dim_country)} baris (C001-C{n})')

# DimYear
all_years = sorted(set(
    long_mortality['tahun'].unique().tolist() +
    long_life['tahun'].unique().tolist() +
    long_pop['tahun'].unique().tolist() +
    df_income_level['tahun'].unique().tolist() +
    df_cause_of_death['tahun'].unique().tolist()
))
df_dim_year = pd.DataFrame({
    'year_id': [f'YR{str(i).zfill(3)}' for i in range(1, len(all_years) + 1)],
    'tahun': all_years
})
print(f'DimYear: {len(df_dim_year)} baris ({list(df_dim_year["year_id"])})')

# DimIncomeLevel
df_dim_income = (
    df_income_level[['level_pendapatan']]
    .drop_duplicates('level_pendapatan')
    .reset_index(drop=True)
)
n = len(df_dim_income)
df_dim_income['income_level_id'] = [f'IL{str(i).zfill(3)}' for i in range(1, n + 1)]
df_dim_income = df_dim_income[['income_level_id', 'level_pendapatan']]
print(f'DimIncomeLevel: {len(df_dim_income)} baris (IL001-IL{n})')
```

```python
# DimCause
df_dim_cause = (
    df_cause_of_death[['penyebab_kematian']]
    .drop_duplicates('penyebab_kematian')
    .reset_index(drop=True)
)
n = len(df_dim_cause)
df_dim_cause['cause_id'] = [f'CS{str(i).zfill(5)}' for i in range(1, n + 1)]
df_dim_cause = df_dim_cause[['cause_id', 'penyebab_kematian']]
print(f'DimCause: {len(df_dim_cause)} baris (CS00001-CS{n})')

# DimSex
df_dim_sex = (
    df_cause_of_death[['jenis_kelamin']]
    .drop_duplicates('jenis_kelamin')
    .reset_index(drop=True)
)
n = len(df_dim_sex)
df_dim_sex['sex_id'] = [f'SX{str(i).zfill(3)}' for i in range(1, n + 1)]
df_dim_sex = df_dim_sex[['sex_id', 'jenis_kelamin']]
print(f'DimSex: {len(df_dim_sex)} baris (SX001-SX{n})')

# DimAgeGroup
df_dim_age = (
    df_cause_of_death[['kelompok_umur']]
    .drop_duplicates('kelompok_umur')
    .reset_index(drop=True)
)
n = len(df_dim_age)
df_dim_age['age_group_id'] = [f'AG{str(i).zfill(3)}' for i in range(1, n + 1)]
df_dim_age = df_dim_age[['age_group_id', 'kelompok_umur']]
print(f'DimAgeGroup: {len(df_dim_age)} baris (AG001-AG{n})')
```

```python
# 3. FACT TABLES

# FactCountryYear: gabung unpivot + income level
print('FactCountryYear:')

fact_cy = long_mortality.merge(
    long_life[['kode_negara', 'tahun', 'life_expectancy']],
    on=['kode_negara', 'tahun'], how='inner'
).merge(
    long_pop[['kode_negara', 'tahun', 'population']],
    on=['kode_negara', 'tahun'], how='inner'
).merge(
    df_income_level[['kode_negara', 'tahun', 'gdp_per_capita', 'level_pendapatan']],
    on=['kode_negara', 'tahun'], how='left'
)

# Map FK
country_map = dict(zip(df_dim_country['kode_negara'], df_dim_country['country_id']))
year_map = dict(zip(df_dim_year['tahun'], df_dim_year['year_id']))
income_map = dict(zip(df_dim_income['level_pendapatan'], df_dim_income['income_level_id']))

fact_cy['country_id'] = fact_cy['kode_negara'].map(country_map)
fact_cy['year_id'] = fact_cy['tahun'].map(year_map)
fact_cy['income_level_id'] = fact_cy['level_pendapatan'].map(income_map)
before = len(fact_cy)
fact_cy = fact_cy.dropna(subset=['income_level_id']).reset_index(drop=True)
n = len(fact_cy)
fact_cy.insert(0, 'fact_country_year_id', [f'FCY{str(i).zfill(5)}' for i in range(1, n + 1)])
fact_cy = fact_cy[['fact_country_year_id', 'country_id', 'year_id', 'income_level_id',
                    'mortality_rate', 'life_expectancy', 'population', 'gdp_per_capita']]
print(f'  Drop aggregate tanpa income_level: {before} -> {n} baris')
print(f'  {n} baris (FCY00001-FCY{n})')
print(f'  Columns: {list(fact_cy.columns)}')
```

```python
# FactDeath
print('FactDeath:')

cause_map = dict(zip(df_dim_cause['penyebab_kematian'], df_dim_cause['cause_id']))
sex_map = dict(zip(df_dim_sex['jenis_kelamin'], df_dim_sex['sex_id']))
age_map = dict(zip(df_dim_age['kelompok_umur'], df_dim_age['age_group_id']))

fact_death = df_cause_of_death.copy()
fact_death['country_id'] = fact_death['kode_negara'].map(country_map)
fact_death['year_id'] = fact_death['tahun'].map(year_map)
fact_death['cause_id'] = fact_death['penyebab_kematian'].map(cause_map)
fact_death['sex_id'] = fact_death['jenis_kelamin'].map(sex_map)
fact_death['age_group_id'] = fact_death['kelompok_umur'].map(age_map)

n = len(fact_death)
fact_death.insert(0, 'fact_death_id', [f'FD{str(i).zfill(7)}' for i in range(1, n + 1)])
fact_death = fact_death[['fact_death_id', 'country_id', 'year_id', 'tahun', 'cause_id', 'sex_id',
                        'age_group_id', 'jumlah_kematian']]
print(f'  {len(fact_death)} baris (FD0000001-FD{n})')
print(f'  Columns: {list(fact_death.columns)}')
```

```python
DDL_SQL = '''

CREATE EXTENSION IF NOT EXISTS pg_stat_statements;

CREATE TABLE DimCountry (
    country_id VARCHAR(6) PRIMARY KEY,
    kode_negara VARCHAR(10),
    negara VARCHAR(100)
);

CREATE TABLE DimYear (
    year_id VARCHAR(6) PRIMARY KEY,
    tahun INTEGER UNIQUE
);

CREATE TABLE DimIncomeLevel (
    income_level_id VARCHAR(6) PRIMARY KEY,
    level_pendapatan VARCHAR(50)
);

CREATE TABLE DimCause (
    cause_id VARCHAR(7) PRIMARY KEY,
    penyebab_kematian VARCHAR(255)
);

CREATE TABLE DimSex (
    sex_id VARCHAR(6) PRIMARY KEY,
    jenis_kelamin VARCHAR(20)
);

CREATE TABLE DimAgeGroup (
    age_group_id VARCHAR(6) PRIMARY KEY,
    kelompok_umur VARCHAR(100)
);

CREATE TABLE FactCountryYear (
    fact_country_year_id VARCHAR(8) PRIMARY KEY,
    country_id VARCHAR(6) REFERENCES DimCountry(country_id),
    year_id VARCHAR(6) REFERENCES DimYear(year_id),
    income_level_id VARCHAR(6) REFERENCES DimIncomeLevel(income_level_id),
    mortality_rate NUMERIC,
    life_expectancy NUMERIC,
    population NUMERIC,
    gdp_per_capita NUMERIC
);

CREATE TABLE FactDeath (
    fact_death_id VARCHAR(10),
    country_id VARCHAR(6) REFERENCES DimCountry(country_id),
    year_id VARCHAR(6) REFERENCES DimYear(year_id),
    tahun INTEGER NOT NULL,
    cause_id VARCHAR(7) REFERENCES DimCause(cause_id),
    sex_id VARCHAR(6) REFERENCES DimSex(sex_id),
    age_group_id VARCHAR(6) REFERENCES DimAgeGroup(age_group_id),
    jumlah_kematian INTEGER,

    PRIMARY KEY (fact_death_id, tahun)
) PARTITION BY RANGE (tahun);

CREATE TABLE FactDeath_2017
PARTITION OF FactDeath
FOR VALUES FROM (2017) TO (2018);

CREATE TABLE FactDeath_2018
PARTITION OF FactDeath
FOR VALUES FROM (2018) TO (2019);

CREATE TABLE FactDeath_2019
PARTITION OF FactDeath
FOR VALUES FROM (2019) TO (2020);

CREATE TABLE FactDeath_2020
PARTITION OF FactDeath
FOR VALUES FROM (2020) TO (2021);

CREATE INDEX idx_fd_country
ON FactDeath(country_id);

CREATE INDEX idx_fd_tahun
ON FactDeath(tahun);

CREATE INDEX idx_fd_cause
ON FactDeath(cause_id);

CREATE INDEX idx_fd_sex
ON FactDeath(sex_id);

CREATE INDEX idx_fd_agegroup
ON FactDeath(age_group_id);

CREATE INDEX idx_fd_country_year
ON FactDeath(country_id, tahun);

CREATE INDEX idx_fd_cause_year
ON FactDeath(cause_id, tahun);

CREATE INDEX idx_fcy_country
ON FactCountryYear(country_id);

CREATE INDEX idx_fcy_year
ON FactCountryYear(year_id);

CREATE INDEX idx_fcy_country_year
ON FactCountryYear(country_id, year_id);

CREATE MATERIALIZED VIEW mv_deaths_by_country_year AS
SELECT
    fd.country_id,
    c.negara,
    fd.tahun,
    SUM(fd.jumlah_kematian) AS total_deaths
FROM FactDeath fd
JOIN DimCountry c
    ON c.country_id = fd.country_id
GROUP BY
    fd.country_id,
    c.negara,
    fd.tahun;

CREATE INDEX idx_mv_country_year
ON mv_deaths_by_country_year(country_id, tahun);

CREATE MATERIALIZED VIEW mv_deaths_by_cause AS
SELECT
    dc.penyebab_kematian,
    fd.tahun,
    SUM(fd.jumlah_kematian) AS total_deaths
FROM FactDeath fd
JOIN DimCause dc
    ON dc.cause_id = fd.cause_id
GROUP BY
    dc.penyebab_kematian,
    fd.tahun;

CREATE INDEX idx_mv_cause_year
ON mv_deaths_by_cause(penyebab_kematian, tahun);

CREATE MATERIALIZED VIEW mv_deaths_by_age AS
SELECT
    da.kelompok_umur,
    fd.tahun,
    SUM(fd.jumlah_kematian) AS total_deaths
FROM FactDeath fd
JOIN DimAgeGroup da
    ON da.age_group_id = fd.age_group_id
GROUP BY
    da.kelompok_umur,
    fd.tahun;

CREATE INDEX idx_mv_age
ON mv_deaths_by_age(kelompok_umur);

CREATE INDEX idx_mv_age_year
ON mv_deaths_by_age(kelompok_umur, tahun);
'''
```

```python
REFRESH_DDL = '''
REFRESH MATERIALIZED VIEW mv_deaths_by_country_year;
REFRESH MATERIALIZED VIEW mv_deaths_by_cause;
REFRESH MATERIALIZED VIEW mv_deaths_by_age;
'''
```

```python
RESET_SQL = """
DROP MATERIALIZED VIEW IF EXISTS mv_deaths_by_country_year;
DROP MATERIALIZED VIEW IF EXISTS mv_deaths_by_cause;
DROP MATERIALIZED VIEW IF EXISTS mv_deaths_by_age;

DROP TABLE IF EXISTS factdeath CASCADE;
DROP TABLE IF EXISTS factdeath_2017 CASCADE;
DROP TABLE IF EXISTS factdeath_2018 CASCADE;
DROP TABLE IF EXISTS factdeath_2019 CASCADE;
DROP TABLE IF EXISTS factdeath_2020 CASCADE;
DROP TABLE IF EXISTS factcountryyear CASCADE;

DROP TABLE IF EXISTS dimcountry CASCADE;
DROP TABLE IF EXISTS dimyear CASCADE;
DROP TABLE IF EXISTS dimincomelevel CASCADE;
DROP TABLE IF EXISTS dimcause CASCADE;
DROP TABLE IF EXISTS dimsex CASCADE;
DROP TABLE IF EXISTS dimagegroup CASCADE;
"""
```

```python
import psycopg2
from psycopg2.extras import execute_values

def insert_df(cur, table, df, cols):
    """Bulk insert DataFrame rows into a PostgreSQL table."""
    rows = [tuple(row) for row in df[cols].values]
    sql = f"INSERT INTO {table} ({', '.join(cols)}) VALUES %s"
    execute_values(cur, sql, rows, template=None, page_size=1000)
    print(f"  {table}: {len(rows)} baris ditambahkan")

conn = psycopg2.connect(
    host=DB_CONFIG['host'],
    port=DB_CONFIG['port'],
    dbname=DB_CONFIG['dbname'],
    user=DB_CONFIG['user'],
    password=DB_CONFIG['password']
)
conn.autocommit = True
cur = conn.cursor()

print('Drop tables:')
cur.execute(RESET_SQL)
print('selesai')

print('\nCreate tables:')
cur.execute(DDL_SQL)
print('selesai')

print('\nInsert dimension:')
insert_df(cur, 'DimCountry', df_dim_country, ['country_id', 'kode_negara', 'negara'])
insert_df(cur, 'DimYear', df_dim_year, ['year_id', 'tahun'])
insert_df(cur, 'DimIncomeLevel', df_dim_income, ['income_level_id', 'level_pendapatan'])
insert_df(cur, 'DimCause', df_dim_cause, ['cause_id', 'penyebab_kematian'])
insert_df(cur, 'DimSex', df_dim_sex, ['sex_id', 'jenis_kelamin'])
insert_df(cur, 'DimAgeGroup', df_dim_age, ['age_group_id', 'kelompok_umur'])

print('\nInsert fact:')
n0 = len(fact_cy)
fact_cy_clean = fact_cy.dropna(subset=['country_id', 'year_id', 'income_level_id']).reset_index(drop=True)
dropped = n0 - len(fact_cy_clean)
if dropped:
    print(f'  {dropped} baris FactCountryYear di-drop')
insert_df(cur, 'FactCountryYear', fact_cy_clean,
            ['fact_country_year_id', 'country_id', 'year_id', 'income_level_id',
            'mortality_rate', 'life_expectancy', 'population', 'gdp_per_capita'])

n0 = len(fact_death)
fact_death_clean = fact_death.dropna(subset=['country_id', 'year_id', 'cause_id', 'sex_id', 'age_group_id']).reset_index(drop=True)
dropped = n0 - len(fact_death_clean)
if dropped:
    print(f'  {dropped} baris FactDeath di-drop')
insert_df(cur, 'FactDeath', fact_death_clean,
            ['fact_death_id', 'country_id', 'year_id', 'tahun', 'cause_id',
            'sex_id', 'age_group_id', 'jumlah_kematian'])

print('\nRefresh MV:')
cur.execute(REFRESH_DDL)
print('selesai.')

cur.close()
conn.close()
print('\nload selesai')
```

```python
import atoti as tt

session = tt.Session.start()
```

```python
df_dim_country = df_dim_country.astype({
    "country_id": "string",
    "kode_negara": "string",
    "negara": "string"
})

df_dim_year = df_dim_year.astype({
    "year_id": "string",
    "tahun": "string"
})

df_dim_income = df_dim_income.astype({
    "income_level_id": "string",
    "level_pendapatan": "string"
})

df_dim_cause = df_dim_cause.astype({
    "cause_id": "string",
    "penyebab_kematian": "string"
})

df_dim_sex = df_dim_sex.astype({
    "sex_id": "string",
    "jenis_kelamin": "string"
})

df_dim_age = df_dim_age.astype({
    "age_group_id": "string",
    "kelompok_umur": "string"
})

fact_cy_clean = fact_cy_clean.astype({
    "fact_country_year_id": "string",
    "country_id": "string",
    "year_id": "string",
    "income_level_id": "string",
    "mortality_rate": "float64",
    "life_expectancy": "float64",
    "population": "float64",
    "gdp_per_capita": "float64"
})

fact_death_clean = fact_death_clean.astype({
    "fact_death_id": "string",
    "country_id": "string",
    "year_id": "string",
    "cause_id": "string",
    "sex_id": "string",
    "age_group_id": "string",
    "jumlah_kematian": "int64"
})

fact_death_clean.drop(columns=["tahun"], inplace=True)
```

```python
country_tbl = session.read_pandas(df_dim_country, table_name='country', keys=["country_id"])
year_tbl = session.read_pandas(df_dim_year, table_name='year', keys=["year_id"])
income_tbl = session.read_pandas(df_dim_income, table_name='income', keys=["income_level_id"])
cause_tbl = session.read_pandas(df_dim_cause, table_name='cause', keys=["cause_id"])
sex_tbl = session.read_pandas(df_dim_sex, table_name='sex', keys=["sex_id"])
age_tbl = session.read_pandas(df_dim_age, table_name='age', keys=["age_group_id"])

fact_country_year_tbl = session.read_pandas(fact_cy_clean, table_name='country_year', keys=["country_id", "year_id"])
fact_death_tbl = session.read_pandas(fact_death_clean, table_name='death', keys=["country_id", "year_id", "cause_id", "sex_id", "age_group_id"])

print('selesai')
```

```python
#  DeathCube 
cube_death = session.create_cube(fact_death_tbl, "DeathCube")

m_death = cube_death.measures
m_death["Total Deaths"] = tt.agg.sum(fact_death_tbl["jumlah_kematian"])

#  CountryYearCube 
cube_country = session.create_cube(fact_country_year_tbl, "CountryYearCube")

m_country = cube_country.measures
m_country["Avg Life Expectancy"] = tt.agg.mean(fact_country_year_tbl["life_expectancy"])
m_country["Avg Mortality Rate"] = tt.agg.mean(fact_country_year_tbl["mortality_rate"])
m_country["Total Population"] = tt.agg.sum(fact_country_year_tbl["population"])
m_country["Avg GDP per Capita"] = tt.agg.mean(fact_country_year_tbl["gdp_per_capita"])

# Ringkasan
print("Cubes:")
print(f"DeathCube: {len(list(cube_death.hierarchies))} hierarchies, {len(list(m_death))} measures")
print(f"CountryYearCube: {len(list(cube_country.hierarchies))} hierarchies, {len(list(m_country))} measures")
```

## Benchmark

```python
import psycopg2, time
conn = psycopg2.connect(
    host=DB_CONFIG['host'], port=DB_CONFIG['port'],
    dbname=DB_CONFIG['dbname'], user=DB_CONFIG['user'],
    password=DB_CONFIG['password']
)
conn.autocommit = True
cur = conn.cursor()

print('Benchmark:')

start = time.time()
cur.execute("SELECT SUM(jumlah_kematian) FROM FactDeath;")
full = cur.fetchone()[0]
t_full = time.time() - start
print(f'  Full scan (semua partisi): {t_full:.5f}s -> {full}')

start = time.time()
cur.execute("SELECT SUM(jumlah_kematian) FROM FactDeath WHERE tahun = 2017;")
part = cur.fetchone()[0]
t_part = time.time() - start
print(f'  Partition pruning (tahun=2017): {t_part:.5f}s -> {part}')

ratio = round(t_full / t_part, 2) if t_part else float('inf')
print(f'  Speedup: {ratio}x')
results = [('Partition Pruning vs Full Scan', t_full, t_part, ratio)]
```

```python
print('\nBenchmark MV:')

start = time.time()
cur.execute("""
    SELECT c.negara, SUM(fd.jumlah_kematian)
    FROM FactDeath fd
    JOIN DimCountry c ON c.country_id = fd.country_id
    GROUP BY c.negara
""")
live = cur.fetchall()
t_live = time.time() - start
print(f'  Live JOIN + GROUP BY: {t_live:.5f}s')

start = time.time()
cur.execute("SELECT * FROM mv_deaths_by_country_year;")
mv = cur.fetchall()
t_mv = time.time() - start
print(f'  Materialized View: {t_mv:.5f}s')

ratio = round(t_live / t_mv, 2) if t_mv else float('inf')
print(f'  Speedup: {ratio}x')
results.append(('Materialized View vs Live JOIN', t_live, t_mv, ratio))
```

```python
print('\nRingkasan:')
print(f"{'Skenario':<30} {'Full scan (s)':<15} {'Optimasi (s)':<15} {'Speedup':<10}")
for skenario, before, after, speedup in results:
    print(f"{skenario:<30} {before:<15.5f} {after:<15.5f} {speedup}x")
cur.close()
conn.close()
```
